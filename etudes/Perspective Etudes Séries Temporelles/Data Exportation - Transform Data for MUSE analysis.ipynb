```python
import pandas as pd
import numpy as np

from transplant.data.dataset import Dataset
dataset=Dataset()

from transplant.data.learningset import Learningset
learningset = Learningset()
```

```python
train_static_0, test_static_0 = learningset.get_static_filled()
train_dynamic_0, test_dynamic_0 = dataset.get_dynamic()

```

```python
train_static_0['target']=train_static_0['target'].map({1: 1, 0: 2})
test_static_0['target']=test_static_0['target'].map({1: 1, 0: 2})
```

```python
df_train=pd.merge(train_dynamic_0, train_static_0[['id_patient','target']], on='id_patient').fillna(0)
df_test=pd.merge(test_dynamic_0, test_static_0[['id_patient','target']], on='id_patient').fillna(0)
```

```python
len(set(df_train['id_patient']))
```

```python
dict_id_train_to_range={}
j=1
for i in set(df_train['id_patient']): 
    dict_id_train_to_range[i]=j
    j+=1
    
dict_id_test_to_range={}
j=1
for i in set(df_test['id_patient']): 
    dict_id_test_to_range[i]=j
    j+=1
```

```python
df_train['id_patient']=df_train['id_patient'].map(dict_id_train_to_range)
df_test['id_patient']=df_test['id_patient'].map(dict_id_test_to_range)
```

```python
df_train.columns
```

```python
df1_train=df_train[['id_patient','time','target','FiO2','B.I.S','SpO2']]
df1_test=df_test[['id_patient','time','target','FiO2','B.I.S','SpO2']]
```

```python
df1_train['nb'] =1
df1_test['nb'] =1
```

```python
df1_train['Time Stamp']= df1_train.groupby('id_patient')['nb'].cumsum()
df1_test['Time Stamp']= df1_test.groupby('id_patient')['nb'].cumsum()
```

```python
TRAIN=df1_train[['id_patient','Time Stamp','target','FiO2','B.I.S','SpO2']]
TEST=df1_test[['id_patient','Time Stamp','target','FiO2','B.I.S','SpO2']]
```
